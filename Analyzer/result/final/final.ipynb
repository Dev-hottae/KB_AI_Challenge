```python
import pandas as pd
```

```python
final_df = pd.read_csv('/home/lab10/JJC/final.csv')
```

```python
final_df.dropna(inplace=True)
```

```python
final_df['score'] = -final_df['score']
```

```python
norm_df = final_df.describe()
```

```python
final_df
```

```python
norm_df
```

```python
final_df['vix'] = (final_df['vix'] - norm_df['vix']['mean']) / norm_df['vix']['std']
final_df['score'] = (final_df['score'] - norm_df['score']['mean']) / norm_df['score']['std']
```

```python
final_df[['vix','score']].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.plot(final_df['vix'], color='r')
plt.legend('VIX')
sns.scatterplot(final_df.index, final_df['score'])
```
