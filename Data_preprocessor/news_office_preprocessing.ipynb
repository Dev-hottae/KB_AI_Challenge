연합뉴스

```python
yn_path = r'C:\Users\dlagh\Desktop\kb_data\{0}\[{0}]news_연합뉴스.json'.format(keyword)
yn_df = pd.read_json(yn_path).sort_values(by=['date'])
```

```python
# 기사 랜덤 50 추출
yn_df.sample(50)
```

```python
# print(yn_df.loc[253417].url)
# yn_df.loc[253417].text
```

```python
yn_df.shape
```

```python
# 기사형식 가진 기사들만 추출
yn_df = yn_df[yn_df['text'].str.contains("\([가-힣]+=연합[가-힣]+\)[가-힣\s]+=",case=False)]
```

```python
# ~~~ (OO=연합ㅇㅇ) ㅇㅇㅇ 기자 = <cut>
yn_df['text'] = yn_df['text'].str.replace('^[\w\W]*\([가-힣]+=연합[가-힣]+\)[가-힣\s]+=', '')
# 이메일이후 <cut>
yn_df['text'] = yn_df['text'].str.replace('([A-Za-z0-9]+@(yna.))(?!.*([A-Za-z0-9]+@(yna.)).*)[\w\W]+', '').replace('\n',' ').replace('\\',' ').replace('\"', ' ').replace('\r', ' ').replace('\t', ' ').replace('  ', ' ')
# 저작권자(c) 연합뉴스 이후 <cut>
yn_df['text'] = yn_df['text'].str.replace('(저작권자\(c\)[\w\W]+)', '').replace('\n',' ').replace('\\',' ').replace('\"', ' ').replace('\r', ' ').replace('\t', ' ').replace('  ', ' ')
```

```python
# 글자갯수 0자 미만 제거(어차피 깨끗한 기사들임)
yn_df['count_str'] = yn_df['text'].str.count("[\s\S]", re.I)
yn_df = yn_df[yn_df['count_str']>=yn_df['count_str'].quantile(q=0.05)]
yn_df = yn_df[yn_df['count_str']<=yn_df['count_str'].quantile(q=0.95)]
```

```python
yn_df.sample(20)
```

```python
yn_df.describe()
```

```python
# 파일저장
yn_df.to_json('pre_[{0}]news_연합뉴스.json'.format(keyword))
```
